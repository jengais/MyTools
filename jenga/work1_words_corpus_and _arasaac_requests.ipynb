```python
# function get corpus as parameter
# and for each word find it in ARASAAC site
# and print word info like id, meaning and pictogram

def func(x):
    corpus = x
    # each word in corpus
    for w in corpus:
        print('\nNew word request')
        # request in ARASAAC API
        r = requests.get('https://api.arasaac.org/api/pictograms/he/search/' + str(w))
        if(r.ok==True):
            r_dict = r.json()
            # find id of that word
            word_id = r_dict[0]['_id']
            
            # check if there is a meaning
            if('meaning' in r_dict[0]['keywords'][0].keys()):
                word_meaning = r_dict[0]['keywords'][0]['meaning']
                print('"{}" Info for word'.format(str(w)))
                print('Word ID:', word_id)
                print('"{}" :Word meaning'.format(word_meaning))

                url = 'https://static.arasaac.org/pictograms/' + str(word_id) + '/' + str(word_id) + '_500.png'
                print('"{}" Print pictogram of'.format(str(w)))
                IPython.display.Image(url, width = 250)
                
            else:
                print('"{}" Info for word'.format(str(w)))
                print('Word ID:', word_id)
                print('.Word with no meaning')
                
                url = 'https://static.arasaac.org/pictograms/' + str(word_id) + '/' + str(word_id) + '_500.png'
                print('"{}" Print pictogram of'.format(str(w)))
                IPython.display.Image(url, width = 250)
        #print if there is no such word 
        else:
            print('"{}" There is no word'.format(str(w)))

# start to check corpus
func(docs_corpus_words)
```
