```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

### JNJ - Johnson & Johnson
### JPM - JPMorgan Chase & Co.

```python
jnj = pd.read_csv("JNJ.csv")
jpm = pd.read_csv("JPM.csv")

# Basic Average Shares for JNJ and JPM
jnj_basic_average_shares = 2632482092
jpm_basic_average_shares = 2988155355
```

```python
jpm
```

```python
# drop unused columns
jnj = jnj.drop(columns=['Open','High','Low','Adj Close','Volume'])
jnj
```

```python
# drop unused columns
jpm = jpm.drop(columns=['Open','High','Low','Adj Close','Volume'])
jpm
```

```python
# visualize stocks price for both companies
close = jnj['Close'].tolist()
date = jnj['Date'].tolist()
close2 = jpm['Close'].tolist()
date2 = jpm['Date'].tolist()
plt.plot(date, close) 
plt.plot(date2, close2) 
plt.grid( linestyle='--', linewidth=1)
plt.legend(["JNJ", "JPM"], loc ="lower right", markerfirst = True, numpoints=1)
plt.xlabel('Date')
plt.ylabel('Close price in USD')
plt.xticks(date[::90])
plt.title('JNJ and JPM prices last 3 years')
plt.style.use('default')
plt.gcf().autofmt_xdate()
plt.show()
```
