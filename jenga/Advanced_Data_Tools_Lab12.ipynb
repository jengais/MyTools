```python
def convert_tweet_to_tuple(tweet):
    screen_name = tweet.user.screen_name
    full_name = tweet.author.name
    
    #created_at = tweet.created_at
    retweet_count = tweet.retweet_count
    
    followers_count = tweet.user.followers_count
    screen_name = tweet.user.screen_name
    content = tweet.text
    description = tweet.user.description
#     created_at = tweet.created_at
#     created_at = tweet.created_at
    
    
    post_tuple = (screen_name, 
                  full_name,
                  followers_count, 
                  description,
                  #created_at,
                  content,
                  retweet_count)
    
    return post_tuple
```

```python
post_tuples = []

for tweet in tweets:
    post_tuple = convert_tweet_to_tuple(tweet)
    post_tuples.append(post_tuple)
```

```python
tweets_df = pd.DataFrame(post_tuples, columns=['username', 
                                               'full_name', 
                                               'followers', 
                                               'description',
                                               #'date',
                                               'tweet',
                                               'retweets'])
```
