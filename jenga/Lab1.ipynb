```python
def myMinMax(arr, new_min, new_max):
    list1 = arr
    
    for x in range(len(list1)):
        list1[x] = ((list1[x]-min(list1))/(max(list1)-min(list1)))*(new_max-new_min)+new_min

    return list1
```

```python
def myZScore(arr):   
    list1 = arr
    
    x = sum(arr)/len(list1)
    sum1 = 0
    for i in range(len(list1)):
        sum1 = sum1 + (list1[i]-x)**2
        
    s = ((sum1)/len(list1))**0.5
    for i in range(len(list1)):
        list1[i] = (list1[i]-x)/s
    
    return list1
```

```python
def myDecimal(arr):
    list1 = arr
    
    v = 0
    for x in range(len(list1)):
        if(abs(list1[x])>=v):
            v = abs(list1[x])
            
    j = len(str(v))
    for x in range(len(list1)):
        list1[x]=list1[x]/(10**j)
    
    return list1
```

```python
def eqFreq(arr):
    
    numOfElem = int(len(arr)/3) + (len(arr)%3 > 0)
    q = [['Bin1:'], ['Bin2:'], ['Bin3:']]
    
    for i in range(numOfElem):
        q[0].append(arr[i])
        q[1].append(arr[i+numOfElem])
        if((i+numOfElem*2)<=len(arr)-1):    
            q[2].append(arr[i+numOfElem*2])
    n = []
    for r in q:
        for c in r:
            n.append(c)
    print(n)
    
```

```python
def eqWidth(arr):
    
    w = int((max(arr)-min(arr))/3) + ((max(arr)-min(arr))%3>0)
    q = [['Bin1:'], ['Bin2:'], ['Bin3:']]
    
    for i in range(len(arr)):
        if(arr[i]<w):
            q[0].append(arr[i])
        if(arr[i]<w*2 and arr[i]>=w):
            q[1].append(arr[i])
        if(arr[i]<w*3 and arr[i]>=w*2 ):
            q[2].append(arr[i])
        
    n = []
    for r in q:
        for c in r:
            n.append(c)
    print(n)
    
```

```python
def sma(arr, num):
    n = 0
    for i in range(num):
        n += arr[i]

    return n/num
    
```

```python
sma(m, 7)
```

```python
def wma(arr, num):
    n = 0
    x = num
    for i in range(num):
        n += arr[i]*x
        x-=1
    div = (num*(num+1))/2
    return n/div
```

```python
wma(m, 9)
```

```python
def ema(arr, num):
    if(num==1):
        return arr[0]
    
    k = 2/(num+1)
    
    return arr[num-1]*k+ema(arr, num-1)*(1-k)    
```

```python
ema(m, 7)
```

```python
def byMeans(arr):
    arr2=arr
    arr2.sort()
    x=0
    y=0
    z=0
    
    numOfElem = int(len(arr2)/3) + (len(arr2)%3 > 0)
    q = [['Bin1:'], ['Bin2:'], ['Bin3:']]
    
    for i in range(numOfElem):
        q[0].append(arr2[i])
        q[1].append(arr2[i+numOfElem])
        if((i+numOfElem*2)<=len(arr2)-1):    
            q[2].append(arr2[i+numOfElem*2])
            
    for i in range(numOfElem):
        x += (q[0][i+1])
        y += (q[1][i+1])
        if((i+numOfElem*2)<=len(arr2)-1): 
            z += (q[2][i+1])
    
    x /= len(q[0])-1
    y /= len(q[1])-1
    z /= len(q[2])-1
    
    q2=[['Bin1:'], ['Bin2:'], ['Bin3:']]
    for i in range(numOfElem):
        q2[0].append(int(x))
        q2[1].append(int(y))
        if((i+numOfElem*2)<=len(arr2)-1):
            q2[2].append(int(z))
    
    n = []
    for r in q2:
        for c in r:
            n.append(c)
    print(n)
    
```

```python
def byBoundaries(arr):
    arr2=arr
    arr2.sort()
    
    numOfElem = int(len(arr2)/3) + (len(arr2)%3 > 0)
    q = [[], [], []]
    
    for i in range(numOfElem):
        q[0].append(arr2[i])
        q[1].append(arr2[i+numOfElem])
        if((i+numOfElem*2)<=len(arr2)-1):    
            q[2].append(arr2[i+numOfElem*2])
            
    min1 = min(q[0])
    min2 = min(q[1])
    min3 = min(q[2])
    
    max1 = max(q[0])
    max2 = max(q[1])
    max3 = max(q[2])
    
    
    q2=[['Bin1:'], ['Bin2:'], ['Bin3:']]
    for i in range(numOfElem):
        
        if(abs(q[0][i]-min1)<=abs(q[0][i]-max1)):
            q2[0].append(min1)
        else:
            q2[0].append(max1)
            
        if(abs(q[1][i]-min2)<=abs(q[1][i]-max2)):
            q2[1].append(min2)
        else:
            q2[1].append(max2)
              
        if((i+numOfElem*2)<=len(arr2)-1):  
            if(abs(q[2][i]-min3)<=abs(q[2][i]-max3)):
                q2[2].append(min3)
            else:
                q2[2].append(max3)
    
    n = []
    for r in q2:
        for c in r:
            n.append(c)
    print(n)
```
