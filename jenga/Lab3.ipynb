```python
#13

df = pd.read_csv("Desktop/us-marriages-divorces-1867-2014.csv")
marriages = df['Marriages'].tolist()
divorces = df['Divorces'].tolist()
yearlist = df['Year'].tolist()
population = df['Population'].tolist()
marriagesPerCapita=[]
divorcesPerCapita=[]

for i in range(len(yearlist)):
    if yearlist[i]==1900:
        marriagesPerCapita.append(marriages[i]/population[i])
        divorcesPerCapita.append(divorces[i]/population[i])
        
    if yearlist[i]==1950:
        marriagesPerCapita.append(marriages[i]/population[i])
        divorcesPerCapita.append(divorces[i]/population[i])
        
    if yearlist[i]==2000:
        marriagesPerCapita.append(marriages[i]/population[i])
        divorcesPerCapita.append(divorces[i]/population[i])
        

years = ['1900', '1950', '2000']

data = {'years' : years,
        'Marriages'   : marriagesPerCapita,
        'Divorces'   : divorcesPerCapita}

source = ColumnDataSource(data=data)

p = figure(x_range=years, y_range=(0, 0.015), title="Number of marriages and divorces per capita in U.S.", tools="")

p.vbar(x=dodge('years', -0.1, range=p.x_range), top='Marriages', width=0.2, source=source,
       color="steelblue", legend_label="Marriages")

p.vbar(x=dodge('years',  0.1,  range=p.x_range), top='Divorces', width=0.2, source=source,
       color="red", legend_label="Divorces")

p.legend.location = "top_right"
p.legend.orientation = "vertical"

show(p)
```

```python
#EX14

df = pd.read_csv("Desktop/actor_kill_counts.csv")
actors = df['Actor'].tolist()
kills = df['Count'].tolist()

pair = ()
for i in range(len(actors)):
    pair+=(actors[i], kills[i])

newactors = []
kills.sort()
for i in range(len(kills)):
    for j in range(len(pair)):
        if kills[i]==pair[j]:
            newactors.append(pair[j-1])

plot = figure(title="The Deadliest actors" , y_range = newactors)
plot.hbar(newactors, left = 0, right = kills, height=0.5, color="steelblue")

show(plot)
```

```python
#EX15

df = pd.read_csv("Desktop/roman-emperor-reigns.csv")
emperors = df['Emperor'].tolist()
cause = df['Cause_of_Death'].tolist()

numEmp = len(emperors)
numAs=0

for i in range(len(cause)):
    if cause[i]=='Assassinated':
        numAs+=1

x = {'Not Assasinated': (numEmp-numAs)/numEmp,'Assasinated': numAs/numEmp}

data = pd.Series(x).reset_index(name='value').rename(columns={'index':'country'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['color'] = ['steelblue', 'red']

p = figure(title="Roman Emperors percentage", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.5,start_angle=cumsum('angle', include_zero=True), 
        end_angle=cumsum('angle'),
        fill_color='color', legend_field='country', source=data)

data["value"] = data['value'].astype(str)
data["value"] = data["value"].str.pad(35, side = "left")
source = ColumnDataSource(data)

labels = LabelSet(x=0, y=1, text='value', angle=cumsum('angle', include_zero=True), source=source)

p.add_layout(labels)
p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None

show(p)
```

```python
#EX16

df = pd.read_csv("Desktop/arcade-revenue-vs-cs-doctorates.csv")
year = df['Year'].tolist()
totalrevenue = df['Total Arcade Revenue (billions)'].tolist()
csdoc = df['Computer Science Doctorates Awarded (US)'].tolist()
relationship=[]

for i in range(len(year)):
    relationship.append(totalrevenue[i]*pow(10,9)/csdoc[i])
    
df = pd.DataFrame()
df['year'] = year
df['relationship'] = relationship

sns.lmplot('year', 'relationship', data=df)

plt.title('Relationship between revenue and number of awarded')
plt.xlabel('Years')
plt.ylabel('Relationship')
```
