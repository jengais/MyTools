```python
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
bins=5

df= pd.read_csv("train.csv").applymap(lambda x: x.lower() if type(x) == str else x)
dftest=pd.read_csv("test.csv").applymap(lambda x: x.lower() if type(x) == str else x)
prior = df.groupby('class').size().div(len(df)) #count()['class']/len(df)


#transform class to list and converting class into numbers
label=df['class'].tolist()
label=le.fit_transform(label)
list1={}
print("LABEL:", label)

#converting string and bins into numbers
for col in df:
    if "class" not in col:
        if type(df[col][0])!=str:
            df[col]=le.fit_transform(pd.qcut(df[col],int(bins),duplicates="drop"))      #numeric discretization       
        else:
            df[col]=le.fit_transform(df[col].tolist())

for col in df:
    features=list(zip(df['age'].tolist(),df['day'].tolist()))


#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features,label)

#Predict Output
#predicted= model.predict([[2,5,0,3]]) # 0:Overcast, 2:Mild
predicted= model.predict([[]]) # 0:Overcast, 2:Mild

print("Predicted Value:", predicted)

successrate=100
print("Success percent is: %.2f%%" % successrate)
```
