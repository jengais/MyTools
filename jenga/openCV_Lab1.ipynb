### TASK 3

```python
image = cv2.imread("FamilyTrip.jpg", cv2.IMREAD_COLOR)
```

```python
resized = cv2.resize(image, (0,0), fx=0.5, fy=0.5)
```

```python
h, w = image.shape[:2]
print("Height = {},  Width = {}".format(h, w))
h, w = resized.shape[:2]
print("Resized Height = {},  Resized Width = {}".format(h, w))
```

```python
image_fixed = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
plt.imshow(image_fixed)
plt.show()
```

```python
cv2.imwrite('Desktop\FamilyTrip_resized.jpg', cv2.cvtColor(image_fixed, cv2.COLOR_RGB2BGR))
```
