```python
df = pd.read_csv("C:\\Users\jenga\Desktop\Stocks\Stocks\AT&T Inc. (T)\T.csv")
df_div = pd.read_csv("C:\\Users\jenga\Desktop\Stocks\Stocks\AT&T Inc. (T)\T_div.csv")
```

```python
df = df.drop(columns=['Open','High','Low','Adj Close','Volume'])
```

```python
df
```

```python
# visualize stocks price for both companies
close = df['Close'].tolist()
date = df['Date'].tolist()
#close2 = jpm['Close'].tolist()
#date2 = jpm['Date'].tolist()
plt.plot(date, close) 
#plt.plot(date2, close2) 
plt.grid( linestyle='--', linewidth=1)
plt.legend(["AT&T"], loc ="lower right", markerfirst = True, numpoints=1)
plt.xlabel('Date')
plt.ylabel('Close price in USD')
plt.xticks(date[::90])
plt.title('AT&T  prices last 5 years')
plt.style.use('default')
plt.gcf().autofmt_xdate()
plt.show()
```

```python
# visualize Income data(LTM) for both companies in last three years
df_div = df_div.sort_values(by='Date')
year = df_div['Date'].tolist()
revenue = df_div['Dividends'].tolist()
#earnings = jpm_financials['Income'].tolist()
#plt.bar(year, earnings, width = 0.3, color = 'steelblue', label = 'JPM',align='center')
plt.bar(year, revenue, width = 0.5, color = 'orange', label = 'JNJ',align='edge')
plt.grid( linestyle='--', linewidth=1)
plt.legend(loc = 'upper left', markerfirst = True, numpoints=1)
plt.xlabel('Year')
plt.ylabel('Trailings 12 Months')
plt.xticks(year)
plt.title('Income in 10 Billions USD')
plt.yticks()
plt.style.use('default')
plt.gcf().autofmt_xdate()
plt.show()
```
